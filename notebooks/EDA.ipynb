```python
import requests
from io import BytesIO
from PIL import Image
```

```python
# url = df[df["newspaper_title"] == "Правда/1935/№ 1 (6247)/2.jpg"]["origin_img"].iloc[0]
# url
url = 'http://electrohelp.ebook.nekrasovka.ru/books/newspaper/103866/0-1.jpg'
```

```python
response = requests.get(url)
img = Image.open(BytesIO(response.content))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
import numpy as np

threshold = 220
opacity = 0.5

img_array = np.array(img)
    
white_mask = np.min(img_array, axis=2) > threshold

print(white_mask.mean())

highlight = np.zeros_like(img_array)
highlight[white_mask] = (255, 0, 0)
    
result = img_array.copy()
result[white_mask] = (opacity * highlight[white_mask] + 
                         (1 - opacity) * img_array[white_mask])
```

```python
plt.figure(figsize=(7, 15))
plt.axis("off")
plt.imshow(result)
# plt.savefig("masked.png")
plt.show()
```
